ICS-3 Diffie-Hellman key exchange

```python
# global variables
p = 23
q = 14

# private keys
a = 3
b = 4

# public keys
Ka = pow(q, a) % p
Kb = pow(q, b) % p

# symm keys after public key exchanged
Sa = pow(Kb, a) % p
Sb = pow(Ka, b) % p

print(Sa)

assert(Sa==Sb)
```
