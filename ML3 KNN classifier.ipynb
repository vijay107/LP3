```python
import numpy as np

class kNN:
    def __init__(self, k):
        self.k=k
        self.X = []
        self.y = []
        
    def fit(self, X, y):  ##
        self.X = self.X + X
        self.y=self.y + y
        
    def _distance(self, x, y):
        return (x[0]-y[0])**2 + (x[1]-y[1])**2
    
    def _get_class(self, X):
        dist = []
        for i in range(len(self.X)):
            dist.append( (self._distance(X, self.X[i]), self.y[i]) ) ##
        dist.sort()
        dist=dist[:self.k]
        counts = {}
        for d in dist: ##
            try: counts[d[1]]+=1
            except: counts[d[1]]=1
        return max(counts, key = lambda i: counts[i]) ##
    
    def predict(self, X):
        preds = []
        for x in X:
            preds.append(self._get_class(x))
        return preds
    
    def _get_weighted_class(self, X):
        dist = []
        for i in range(len(self.X)):
            dist.append( (self._distance(X, self.X[i]), self.y[i]) ) 
        dist.sort()
        dist=dist[:self.k]
        counts = {}
        for d in dist: ##
            try: counts[d[1]] += 1/d[0]
            except: counts[d[1]] = 1/d[0]
        return max(counts, key = lambda i: counts[i])
    
    def predict_weighted_class(self, X):
        preds = []
        for x in X:
            preds.append(self._get_weighted_class(x))
        return preds
    
    def _get_locally_weighted_avg_class(self, X):
        dist = []
        for i in range(len(self.X)):
            dist.append( (self._distance(X, self.X[i]), self.y[i]) ) 
        dist.sort()
        dist=dist[:self.k]
        counts = {}
        for d in dist: ##
            try: counts[d[1]].append(1/d[0])
            except: counts[d[1]] = 1/d[0]
        for c in counts: ##
            counts[c]=np.mean(counts[c]) ##
        return max(counts, key = lambda i : counts[i])
    
    def predict_locally_weighted_avg_class(self, X):
        preds = []
        for x in X:
            preds.append(self._get_locally_weighted_avg_class(x))
        return preds
    
X = [
     (2, 4),
     (4, 6),
     (4, 4),
     (4, 2),
     (6, 4),
     (6 ,2)
]
y = ['Y', 'Y', 'B', 'Y', 'Y', 'B']

model = kNN(3)

model.fit(X, y)

print(f'Standard k-NN: {model.predict([(6, 6)])}')      

print(f'wighted k-NN: {model.predict_weighted_class([(6,6)])}')

print(f'locally wighted avg k-NN: {model.predict_locally_weighted_avg_class([(6,6)])}')
```
