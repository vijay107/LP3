```python
## helper functions
def gcd(a, b):   ## or  use math.gcd(a, b)
    if a == 0:
        return b
    return gcd(b % a, a)

def mod_pow(a, b, n):
    return pow(a, b) % n

p = 53
q = 59
n = p*q
pn = (p-1)*(q-1)

## generating keys

# Generating e
e = 2
while e < pn:
    if gcd(e, pn) == 1:
        break
    e += 1

# Generating d    mug up
k = 1
while (k * pn + 1) % e != 0:
    k += 1
d = (k * pn + 1) //e

# keys
pb = [e, n]
pr = [d, n]

print('primes = ', p , ", ", q)
print('n = ', n)
print('pn = ', pn)
print('e = ', e)
print('d =', d)
print('Public key = ', pb)
print('private key = ', pr)

## testing
def encrypt(pt, pb):
    e, n = pb
    ct = mod_pow(pt , e, n)
    return ct

def decrypt(ct, pr):
    d, n = pr
    pt = mod_pow(ct, d, n)
    return pt 

ptxt = 89

c = encrypt(ptxt, pb)

ans = decrypt(c, pr)

assert(ans == ptxt)
```
