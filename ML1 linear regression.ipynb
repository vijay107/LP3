```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set( rc = {'figure.figsize':(8,8)} )  ###

data=[
    (10, 95),
    (9, 80),
    (2, 10),
    (15, 50),
    (10, 45),
    (16, 98),
    (11, 38),
    (16, 93),
]

## step 1-> 4 arrays and their sum
n = len(data)
x = [pt[0] for pt in data]
y = [pt[1] for pt in data]
xx = [a*a for a in x]
xy = [x[i]*y[i] for i in range(n)]

#sns.scatterplot(x=x, y=y, s=100)

sum_x = np.sum(x)
sum_y = np.sum(y)
sum_xx = np.sum(xx)
sum_xy = np.sum(xy)


##step 2-> slope and y intercept
#slope
m = ( n*sum_xy - sum_x*sum_y )/( n*sum_xx - sum_x*sum_x)

#y-intercept -> y = mx + c   here c is y intercept
c = (sum_y - m*sum_x)/n     #remember to divide by n

print(f'Linear Equation is: y = {round(m,2)}*x + {round(c,2)}')

def plot_g(x, y, slope, intercept):
    axes=sns.scatterplot(x=x, y=y, s=100) ###
    x_val=np.array(axes.get_xlim()) ###
    y_val=slope*x_val + intercept
    plt.plot(x_val, y_val, color='red')

plot_g(x, y, m, c)
```
