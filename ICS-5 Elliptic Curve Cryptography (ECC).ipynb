```python
import math
import random

## 1->generates genartor point from given point, generator is used for public keys and c1 
def point(a, b):
 if (4*(a**3) + 27*(b**2)) != 0: ###
  x = 1
  while True:
    rhs = (x**3) + (a*x) + b ###
    y = int(math.sqrt(rhs))
    lhs = (y**2)

    if lhs == rhs:
      return [x, y]

    else:
      x += 1

 else:
  print("Enter another coefficients.")
```

```python
# 2->input points
a = int(input("Enter the coefficient 'a' of curve: "))
b = int(input("Enter the coefficient 'b' of curve: "))
```

```python
## 3->initialize private keys and input plaintext
private_A = 13
private_B = 6
generator = point(a, b)
print("Generator point: ", generator)
m = int(input("Enter the plaintext integer: "))
```

```python
## 4->find public keys , generate random number k(0, 10), find ciphertext[c1, c2]
## and find plaintext from ciphertext    (k is needed for c1 and c2)
public_key_A = [private_A*generator[0], private_A*generator[1]]
print("Public Key of A: ", public_key_A)

public_key_B = [private_B*generator[0], private_B*generator[1]]
print("Public Key of B: ", public_key_B)  ##public keys used in c2

k = random.randint(0, 10)  ## used for c1 and c2
#print(k)
c1 = k*generator[0] + k*generator[1] 
c2 = m + ((k*public_key_B[0]) + (k*public_key_B[1])) ###
ciphertext = [c1, c2]
print("Ciphertext: ", ciphertext)
plaintext = c2 - (c1*private_B)
print("Decrypted Plaintext: ", plaintext)
```
